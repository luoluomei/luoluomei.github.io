```python
import os
import subprocess
import pandas as pd
import time
import random

def run_attack(dataset: str,
               shadow_models: list,
               attack_ids: list,
               seed_num: int = 5,
               gpu: int = 0):
    """
    Run multiple attack experiments for given dataset, shadow model(s), and attack index(es).

    Parameters:
    - dataset: e.g., "cora_ml", "lastfm"
    - shadow_models: list of models like ["graphsage", "gin"]
    - attack_ids: list of attack ids (0-9)
    - seed_num: number of random seeds
    - gpu: GPU device ID
    """
    allowed_models = ["graphsage", "gin", "gat", "gcn"]
    assert all(m in allowed_models for m in shadow_models), f"Allowed models: {allowed_models}"
    assert all(a in range(10) for a in attack_ids), "attack_ids must be between 0 and 9"

    model_dir = "./data/save_model/gnn"
    os.makedirs(model_dir, exist_ok=True)
    log_path = "/content/link_steal_pets2024/output/logs/attack_performance.txt"
    result_dir = "/content/link_steal_pets2024/output/results"
    os.makedirs(result_dir, exist_ok=True)

    # Always use GraphSAGE as target model
    target_pth = os.path.join(model_dir, f"inductive_{dataset}_graphsage_target.pth")
    if not os.path.exists(target_pth):
        print(f"Training target model for {dataset}")
        subprocess.run([
            "python", "train_gnn.py",
            "--dataset", dataset,
            "--model", "graphsage",
            "--mode", "target",
            "--gpu", str(gpu)
        ], check=True)
    else:
        print(f"Target model already exists: {target_pth}")

    attack_args_map = {
        0: ["--node_topology", "0-hop"],
        1: ["--node_topology", "1-hop"],
        2: ["--node_topology", "2-hop"],
        3: ["--node_topology", "0-hop", "--plus"],
        4: ["--node_topology", "1-hop", "--plus"],
        5: ["--node_topology", "2-hop", "--plus"],
        6: ["--node_topology", "1-hop", "--plus2"],
        7: ["--node_topology", "2-hop", "--plus2"],
        8: ["--node_topology", "1-hop", "--all"],
        9: ["--node_topology", "2-hop", "--all"],
    }

    method_map = {
        0: "0-hop_posteriors",
        1: "1-hop_posteriors",
        2: "2-hop_posteriors",
        3: "0-hop_posteriors_node",
        4: "1-hop_posteriors_node",
        5: "2-hop_posteriors_node",
        6: "1-hop_posteriors_graph",
        7: "2-hop_posteriors_graph",
        8: "1-hop_posteriors_node_graph",
        9: "2-hop_posteriors_node_graph",
    }

    seeds = random.sample(range(10000), seed_num)

    all_results = []

    for shadow_model in shadow_models:
        shadow_pth = os.path.join(model_dir, f"inductive_{dataset}_{shadow_model}_shadow.pth")
        if not os.path.exists(shadow_pth):
            print(f"Training shadow model {shadow_model} for {dataset}")
            subprocess.run([
                "python", "train_gnn.py",
                "--dataset", dataset,
                "--model", shadow_model,
                "--mode", "shadow",
                "--gpu", str(gpu)
            ], check=True)
        else:
            print(f"Shadow model already exists: {shadow_pth}")

        for attack_id in attack_ids:
            print(f"\n--- Attack-{attack_id} | Dataset: {dataset} | Shadow: {shadow_model} ---")
            result_path = os.path.join(result_dir, f"attack{attack_id}_summary.csv")

            # Clear log before each attack
            with open(log_path, "w") as f:
                pass

            aucs = []

            for seed in seeds:
                print(f"Running seed {seed}")
                cmd = [
                    "python", "mlp_attack.py",
                    "--dataset", dataset,
                    "--edge_feature", "all",
                    "--target_model", "graphsage",
                    "--shadow_model", shadow_model,
                    "--lr", "0.006",
                    "--optim", "adam",
                    "--scheduler",
                    "--gpu", str(gpu),
                    "--seed", str(seed)
                ] + attack_args_map[attack_id]

                subprocess.run(cmd, check=True)
                time.sleep(1)

                try:
                    with open(log_path, "r") as f:
                        lines = f.readlines()
                    matched = [
                        line for line in lines
                        if dataset in line and str(seed) in line and method_map[attack_id] in line
                    ]
                    if matched:
                        fields = matched[-1].strip().split(",")
                        test_auc = float(fields[11])
                        aucs.append(test_auc)
                        print(f"AUC = {test_auc:.4f}")
                    else:
                        print(f"No match for seed {seed}")
                except Exception as e:
                    print(f"Error extracting AUC for seed {seed}: {e}")

            if aucs:
                avg_auc = round(sum(aucs) / len(aucs), 4)
                print(f"Average AUC: {avg_auc:.4f}")
                df = pd.DataFrame({
                    "dataset": [dataset] * len(aucs) + [dataset],
                    "shadow_model": [shadow_model] * len(aucs) + [shadow_model],
                    "attack_id": [attack_id] * len(aucs) + [attack_id],
                    "seed": seeds + ["avg"],
                    "test_auc": aucs + [avg_auc]
                })
                df.to_csv(result_path, mode="a", index=False, header=not os.path.exists(result_path))
                all_results.append((shadow_model, attack_id, avg_auc))
            else:
                print(f"No AUCs for Attack-{attack_id} with shadow {shadow_model}")

    return all_results
```
